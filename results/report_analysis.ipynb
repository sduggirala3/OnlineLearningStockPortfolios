```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
opening = pd.read_excel('lowvol_daily.xlsx')[::-1].iloc[:,::2]
closing = pd.read_excel('lowvol_daily.xlsx')[::-1].iloc[:,1::2]
opening['BRK.B_open'][opening['BRK.B_open']>500] = 0
#closing['BRK.B_close'].iloc[:185] = 0
#opening['BRK.B_open'].iloc[:172]
```

```python
fig, ax = plt.subplots()
ax.set_title('Low volatility portfolio, daily returns')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.plot(opening)
#ax.set_color_cycle(colors)
plt.xticks(opening.index[::500],rotation=20)
plt.show()
#plt.getp(ax.lines[0].get_color())
```

```python
opening = pd.read_excel('highvol_daily.xlsx')[::-1].iloc[:,::2]
closing = pd.read_excel('highvol_daily.xlsx')[::-1].iloc[:,1::2]
#opening['BRK.B_open'].iloc[:185] = 0
#closing['BRK.B_close'].iloc[:185] = 0
```

```python
fig, ax = plt.subplots()
ax.set_title('High volatility portfolio, daily returns')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.plot(opening)
plt.xticks(opening.index[::500],rotation=20)
plt.show()
```

```python
returns = pd.DataFrame()
for i in range(len(opening.columns)):
    tick = closing.iloc[:,i].name[:4]
    returns[tick]=closing.iloc[:,i]-opening.iloc[:,i]
y =np.std(returns)
```

```python
total_high = (closing.iloc[-1].values - opening.iloc[0].values)/opening.iloc[0].values
total_high
```

Make sure x,y have been set with WEEKLY returns, total_low,total_high have been set with HOURLY

```python
bins = np.linspace(0, 20, 50)

plt.hist(x, bins, alpha=0.5, label='Low volatility')
plt.hist(y, bins, alpha=0.5, label='High volatility',color='green')
plt.legend(loc='upper right')
plt.title('Histogram of stock volatilities by portfolio')
plt.xlabel('Volatility')
plt.ylabel('Count')
plt.show()
```

```python
bins = np.linspace(-.6, .6, 30)

plt.hist(total_low, bins, alpha=0.5, label='Low volatility')
plt.hist(total_high, bins, alpha=0.5, label='High volatility',color='green')
plt.legend(loc='upper right')
plt.title('Histogram of stock returns by portfolio')
plt.xlabel('Total return')
plt.ylabel('Count')
plt.show()
```
